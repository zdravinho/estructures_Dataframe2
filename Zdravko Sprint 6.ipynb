```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score



# Splitting the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('MEDV', axis=1), df['MEDV'], test_size=0.2, random_state=42)

# Creating a Ridge regression model
ridge_model = Ridge(alpha=1.0, fit_intercept=True, normalize=False, copy_X=True, max_iter=None, tol=0.001, solver='auto', random_state=None)

ridge_model.fit(X_train, y_train)

# Making predictions on test set
y_pred = ridge_model.predict(X_test)

# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Dividiendo el conjunto de datos en características y etiquetas
X = df.drop('MEDV', axis=1)
y = df['MEDV']

# Normalizando las características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Creando un modelo de regresión lineal
lin_model = LinearRegression()

# Aplicando la validación cruzada con 5 folds
mse_scores = -1 * cross_val_score(lin_model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')
r2_scores = cross_val_score(lin_model, X_scaled, y, cv=5)

# Imprimiendo los resultados de la validación cruzada

print('Error cuadrático medio:', mse_scores.mean())
print('Coeficiente R2:', r2_scores.mean())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


# Eliminando la variable de número de habitaciones del conjunto de datos
df = df.drop('RM', axis=1)

# Dividiendo el conjunto de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(df.drop('MEDV', axis=1), df['MEDV'], test_size=0.3, random_state=42)

# Creando un modelo de regresión lineal y entrenándolo en el conjunto de entrenamiento
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)

# Haciendo predicciones sobre el conjunto de prueba y evaluando el modelo
y_pred_lin = lin_model.predict(X_test)
mse_lin = mean_squared_error(y_test, y_pred_lin)
r2_lin = r2_score(y_test, y_pred_lin)


print('Error cuadrático medio:', mse_lin)
print('Coeficiente R2:', r2_lin)
```
